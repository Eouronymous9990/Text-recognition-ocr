```python
import cv2
import pytesseract
import matplotlib.pyplot as plt

# Set the path to the Tesseract OCR executable
pytesseract.pytesseract.tesseract_cmd = r"E:\vsc\tesseract.exe"
img = cv2.imread(r"C:\Users\zbook 17 g3\Desktop\New folder (4)\Screenshot 2023-07-22 204722.png")


gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
th3=cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
gaussian=cv2.GaussianBlur(th3, (5,5), 0)


text = pytesseract.image_to_string(gaussian,config="--psm 3")
print(text)

#cv2.imshow("img",th3)
#cv2.waitKey(0)
#cv2.destroyAllWindows() 


rgb_img=cv2.cvtColor(gaussian,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)

```
